```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('menu.csv')
df.head()
```

```python
df.describe()
```

```python
df.fillna(0)
```

```python
graph1 = df.groupby("Category")["Calories"].mean()
graph1 = pd.DataFrame(graph1)
graph1 = graph1.sort_values(by=["Calories"])

print(graph1)
```

```python
fig, ax = plt.subplots()
x = graph1.index
y = graph1["Calories"]
ax.barh(x, y)

bars = ax.barh(x, y, color='lightblue')
for bar, category in zip(bars, x):
    if category == 'Desserts':  # Specify the category you want to highlight
        bar.set_color('red')  # Change to desired color

plt.title("A SOBREMESA Ã© o tipo de refeiÃ§ao com menos calorias no MC Donalds!!")
plt.show()
```
