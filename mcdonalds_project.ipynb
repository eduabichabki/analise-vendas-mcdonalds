```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('menu.csv')
df.head()
```

```python
df.describe()
```

```python
df.fillna(0)
```

```python
calorias_media = pd.DataFrame(df.groupby("Category")["Calories"].mean())
calorias_media = pd.DataFrame(calorias_media)
calorias_media = calorias_media.sort_values(by=["Calories"])

print(calorias_media)
```

```python
fig, ax = plt.subplots()
x = calorias_media.index
y = calorias_media["Calories"]
ax.barh(x, y)

bars = ax.barh(x, y, color='grey')
for bar, category in zip(bars, x):
    if category == 'Desserts':  # Specify the category you want to highlight
        bar.set_color('blue')  # Change to desired color

plt.title("A SOBREMESA Ã© o tipo de refeiÃ§ao com menos calorias no MC Donalds!!")
plt.show()
```

```python
desserts = df[df['Category'] == 'Desserts']
sandwiches = df[df['Category'].isin(['Beef & Pork', 'Chicken & Fish'])]
smoothies = df[df['Category'] == 'Smoothies & Shakes']
```

```python
print("Sobremesas:\n", desserts[['Item', 'Calories']].sort_values('Calories', ascending=False))
```

```python
print("\nLanches:\n", sandwiches[['Item', 'Calories']].sort_values('Calories', ascending=False))
```

```python
print("Sobremesas:\n", smoothies[['Item', 'Calories']].sort_values('Calories', ascending=False))
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x='Category', y='Calories', data=df[df['Category'].isin(['Desserts', 'Beef & Pork', 'Chicken & Fish', 'Smoothies & Shakes', 'Breakfast'])])
plt.title('DistribuiÃ§Ã£o de Calorias: Sobremesas vs. Lanches')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(x='Calories', y='Sugars', hue='Item', data=desserts)
plt.title('Calorias vs. AÃ§Ãºcares em Sobremesas')
plt.show()
```

```python
print("MÃ©dia Calorias de Gordura (Sobremesas):", desserts['Calories from Fat'].mean())
print("MÃ©dia Calorias de Gordura (Lanches):", sandwiches['Calories from Fat'].mean())
```

```python
# Filtrar as categorias desejadas
filtered_df = df[df['Category'].isin(['Desserts', 'Beef & Pork', 'Chicken & Fish', 'Smoothies & Shakes', 'Breakfast'])]

# Definir uma paleta de cores personalizada: azul para 'Desserts', cinza para as demais
colors = ['blue' if cat == 'Desserts' else 'grey' for cat in filtered_df['Category'].unique()]

# Barplot comparativo
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Calories from Fat', hue='Category', data=filtered_df, palette=colors, legend=False)
plt.title('Calorias Provenientes de Gordura por Categoria')
plt.show()
```
